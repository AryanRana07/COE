3. Write a Python program to create a 24-hour time format (HH:MM) using the 4 given digits. Display the latest time
and do not use any digit more than once.

```python
import itertools

def time(digits):

    permutations = itertools.permutations(digits)
    valid_times = filter(is_valid_time, permutations)
    latest_time = max(valid_times)
    formatted_time = "{:02d}:{:02d}".format(latest_time[0], latest_time[1])
    return formatted_time

def is_valid_time(digits):
   
    hours = digits[0] * 10 + digits[1]
    minutes = digits[2] * 10 + digits[3]
    return hours < 24 and minutes < 60


digits = [1, 9, 3, 8]
latest_time = time(digits)
print(latest_time)
```

4. Write a Python program to find the maximum length of a substring in a given string where all the characters of the
substring are the same. Use the itertools module to solve the problem.

```python
from itertools import groupby

def substring(string):
    length = 0
    for i,n in groupby(string):
        length = __builtins__.max(length, len(list(n)))
    return length

string = "abbbbcc"
length = substring(string)
print(length)
```

5. Write a Python program to rotate a Deque Object a specified number (positive) of times. Sample Output: Deque before rotation: deque([2, 4, 6, 8, 10])
Deque after 1 positive rotation: deque([10, 2, 4, 6, 8])
Deque after 2 positive rotations: deque([6, 8, 10, 2, 4])

```python
from collections import deque

def rotate(obj, rt):
    if rt > 0:
        for i in range(rt):
            element = obj.pop()
            obj.appendleft(element)
    elif rt < 0:
        rt = abs(rt)
        for i in range(rt):
            element = obj.popleft()
            obj.append(element)

deque_obj = deque([1,2,3,4,5,6])
print(deque_obj)
rotate(deque_obj, 1)
print(deque_obj)

```

6. Write a Python program to find the majority element from a given array of size n using the Collections module.

```python
from collections import Counter

def major(arr):
    counter = Counter(arr)
    l = len(arr)
    for i,n in counter.items():
        if n > l / 2:
            return i
    return None

a = [2, 4, 5, 6, 6, 6, 6, 6]
element = major(a)
print(element)

```

9. Using Pillow, add a watermark on your image in the middle with a rectangle around it. Watermark should have the
text “This is  a watermarked box”

```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

img = Image.open("Box.jpg")

text = "This is a watermarked box"
font = ImageFont.truetype("arial.ttf",160)

watermark_image = img.copy()
draw = ImageDraw.Draw(watermark_image)
iw,ih = img.size
tw, th = draw.textsize(text, font)

text_x = (iw - tw) // 2
text_y = (ih - th) // 2

rectangle_coords = [(60,750), (1970,1250)]
draw.rectangle(rectangle_coords, outline="black", width=10)
draw.text((text_x, text_y), text, fill="purple", font=font)
plt.imshow(watermark_image)

```

11. Write a Python program to get days between two dates. Sample Dates : 2000,2,28, 2001,2,28
Expected Output : 366 days, 0:00:00

```python
from datetime import datetime, timedelta

d1 = datetime(2001, 9, 1)
d2 = datetime(2023, 6, 2)

diff = d2 - d1

d = diff.days
time_left = diff - timedelta(days=d)

print(d)
print(time_left)
```
