```python
import pandas as pd

# Load the dataset
data = pd.read_csv('winequality-red.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
# Fill missing values with mean
data = data.fillna(data.mean())
```

```python
data.head()
```

```python
# Create a new column for average acidity
data['avg_acidity'] = (data['fixed acidity'] + data['volatile acidity']) / 2
```

```python
data.head()
```

```python
# Delete the columns 'fixed acidity' and 'volatile acidity'
data = data.drop(['fixed acidity', 'volatile acidity'], axis=1)
```

```python
data.head()
```

```python
# Drop duplicate rows
data = data.drop_duplicates()
```
